# turning to grayscale image


```python
gimg=cv2.imread("flower.jpg",0)

```

0 makes image in grayscale


# shape of grayscale image

```python
print(gimg.shape)
# represents row and columns of one single channel.
```

# 2D ARRAY

```python
print(gimg)
```

```python

cv2.imshow("gray",gimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Invert image

```python
In_img=255 - gimg

```

```python

cv2.imshow("invert",In_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Blurr image

```python
blurred = cv2.GaussianBlur(In_img, (21, 21), 0)
```

# resize


```python

cv2.imshow("Blurred",blurred)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
print(blurred)
```

# Invert the blurr image

```python
In_blurr = 255 - blurred
```

```python
sketch= cv2.divide(gimg,In_blurr,scale=256.0)
cv2.imshow("sketch",sketch)
cv2.waitKey(0)
```
